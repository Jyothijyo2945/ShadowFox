```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\satya\ShadowFox\Advanced_Task\Dataset\matches.csv")

df.head()
```

```python
# Dataset Info
df.info()
df.describe()
```

```python
# Check missing values

df.isnull().sum()
```

# Research Questions

1. Which IPL team won the most matches?
2. Which stadium hosted the most IPL matches?
3. Does winning the toss increase winning chances?
4. Which seasons had the highest number of matches?
5. Which teams performed consistently across seasons?

```python
# Top Winning Teams 

plt.figure(figsize=(10,5))
df['winner'].value_counts().head(10).plot(kind='bar')
plt.title("Top Winning Teams")
plt.xlabel("Teams")
plt.ylabel("Wins")
plt.savefig("Images/top_teams.png")
plt.show()
```

```python
# Toss Decision Analysis

plt.figure(figsize=(6,4))
sns.countplot(x='toss_decision', data=df)
plt.title("Toss Decisions")
plt.savefig("Images/toss_decisions.png")
plt.show()
```

```python
# Stadium Analysis

plt.figure(figsize=(10,5))
df['venue'].value_counts().head(10).plot(kind='bar')
plt.title("Top Stadiums")
plt.xlabel("Venue")
plt.ylabel("Matches")
plt.savefig("Images/stadiums.png")
plt.show()
```

```python
# Season Wise Matches

plt.figure(figsize=(10,5))
df['season'].value_counts().sort_index().plot()
plt.title("Season Wise Matches")
plt.xlabel("Season")
plt.ylabel("Matches")
plt.savefig("Images/seasons.png")
plt.show()
```

```python
# Toss Impact Pie Chart

toss_match = (df['toss_winner'] == df['winner']).value_counts()
plt.figure(figsize=(5,5))
plt.pie(
    toss_match,
    labels=['Won Match', 'Lost Match'],
    autopct='%1.1f%%'
)
plt.title("Impact of Toss on Match Result")
plt.savefig("Images/toss_impact.png")
plt.show()
```

# Insights

- Mumbai Indians and Chennai Super Kings are among the top winning teams.
- Most teams prefer fielding after winning the toss.
- Some stadiums host IPL matches more frequently.
- Winning the toss slightly improves winning probability.
- IPL popularity and match count increased across seasons.
