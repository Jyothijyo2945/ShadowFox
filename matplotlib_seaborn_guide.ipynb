```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Sample Dataset

```python
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'Sales': [200, 250, 300, 280, 350],
    'Profit': [20, 30, 40, 35, 50]
}

df = pd.DataFrame(data)

df
```

## Matplotlib Documentation

### 1. Line Plot
   
##### Explanation:

Used to show trends over time.

```python
plt.figure(figsize=(6,5))
plt.plot(df['Month'],df['Sales'])
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.savefig("lineplot.png")
plt.show()
```

### 2. Bar Chart

##### Explanation:

Used to compare categories.

```python
plt.figure(figsize=(6,5))
plt.bar(df['Month'],df['Profit'])
plt.title("Monthly Profit")
plt.xlabel("Month")
plt.ylabel("Profit")
plt.savefig("barplot.png")
plt.show()
```

### 3. Scatter Plot

##### Explanation:

Shows relationship between variables.

```python
plt.figure(figsize=(6,5))
plt.scatter(df['Sales'],df['Profit'])
plt.title("Sales vs Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.savefig("scatterplot.png")
plt.show()
```

### 4. Histogram

##### Explanation:

Shows data distribution.

```python
plt.figure(figsize=(6,5))
plt.hist(df['Sales'])
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.savefig("histplot.png")
plt.show()
```

### 5. Pie Chart

##### Explanation:

Shows percentage contribution.

```python
plt.figure(figsize=(6,5))
plt.pie(df['Sales'], labels=df['Month'], autopct='%1.1f%%')
plt.title("Sales Share")
plt.savefig("pieplot.png")
plt.show()
```

# Seaborn Documentation

### 1. Line Plot

```python
plt.figure(figsize=(6,5))
sns.lineplot(x='Month', y='Sales', data=df)
plt.title("Sales Trend")
plt.savefig("sns_lineplot.png")
plt.show()
```

### 2. Bar Plot

```python
plt.figure(figsize=(6,5))
sns.barplot(x='Month', y='Profit', data=df)
plt.title("Profit Analysis")
plt.savefig("sns_barplot.png")
plt.show()
```

### 3. Scatter Plot

```python
plt.figure(figsize=(6,5))
sns.scatterplot(x='Sales', y='Profit', data=df)
plt.title("Sales vs Profit")
plt.savefig("sns_scatterplot.png")
plt.show()
```

### 4. Histogram Plot

```python
plt.figure(figsize=(6,5))
sns.histplot(df['Sales'])
plt.title("Sales Histogram")
plt.savefig("sns_histogram.png")
plt.show()
```

### 5. Heatmap

```python
plt.figure(figsize=(6,5))
corr = df[['Sales', 'Profit']].corr()
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.savefig("sns_heatmap.png")
plt.show()
```
