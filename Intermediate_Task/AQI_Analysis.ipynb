```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\satya\Downloads/delhiaqi.csv")

df.head()
```

```python
# Dataset Info
df.info()
df.describe()
```

```python
# Checking Missing Values
df.isnull().sum()
```

```python
# Remove Missing Values
df = df.dropna()
```

5. How do pollutants correlate with AQI?

```python
# Correlation Heatmap

plt.figure(figsize=(10,6))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.title("Pollutants Correlation Heatmap")
plt.savefig("images/heatmap.png")
plt.show()
```

```python
# PM2.5 Distribution Histogram

plt.figure(figsize=(8,5))
sns.histplot(df['pm2_5'], bins=20)
plt.title("PM2.5 Distribution")
plt.xlabel("PM2.5")
plt.ylabel("Frequency")
plt.savefig("images/pm25_distribution.png")
plt.show()
```

```python
# Monthly Trend Analysis

sample_data = df.head(50)
plt.figure(figsize=(10,5))
plt.plot(sample_data['pm2_5'])
plt.title("PM2.5 Trend Analysis")
plt.xlabel("Observations")
plt.ylabel("PM2.5")
plt.savefig("images/pm25_trend.png")
plt.show()
```

```python
# Pollutants Comparison

pollutants = ['co', 'no2', 'o3', 'so2', 'pm2_5', 'pm10']
df[pollutants].mean().plot(kind='bar')
plt.title("Average Pollutant Levels")
plt.ylabel("Average Value")
plt.savefig("images/pollutants.png")
plt.show()
```

```python
# Scatter Plot

plt.figure(figsize=(8,5))
sns.scatterplot(x='pm2_5', y='pm10', data=df)
plt.title("PM2.5 vs PM10")
plt.savefig("images/scatterplot.png")
plt.show()
```

# Insights

- PM2.5 and PM10 show strong pollution trends.
- Certain months show higher pollution levels.
- Pollutants are positively correlated with each other.
- PM2.5 contributes significantly to air pollution.

# Conclusion

The analysis shows that PM2.5 and PM10 are major contributors to air pollution. Pollution levels vary across months, and several pollutants are strongly correlated. Data visualization helps in understanding environmental pollution trends effectively.
